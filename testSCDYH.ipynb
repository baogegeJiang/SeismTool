```python
from SeismTool.io import seism
from imp import reload
from obspy import UTCDateTime,imaging
from matplotlib import pyplot as plt
import numpy as np
reload(seism)
quakes0 = seism.QuakeL('sc.meca?')
for quake in quakes0:
    quake['time']=quake['time']-8*3600
quakes1 = seism.QuakeL('sc.meca??')
quakes = seism.QuakeL(quakes0+quakes1[:0])
```

```python
print(UTCDateTime(quakes[0]['time']),quakes[0]['sdr0'],quakes[0]['sdr1'])
print(quakes[-1]['dep'],UTCDateTime(quakes[-1]['time']),quakes[-1]['sdr0'],quakes[-1]['sdr1'])
```

```python
laL=[29,34]
loL=[101,106]
quakes.select({'R':laL+loL})
quakes.write('sc.mecaAll',quakeKeysIn='strTime la lo dep ml strike0 dip0 rake0')
plt.close()
fig=plt.figure(figsize=[10,10])
ax=plt.axes()
plt.xlim(loL)
plt.ylim(laL)
#plt.plot(np.arange(100))
sdr0L=[]
xyL=[]
for quake in quakes[:]:
    #np=imaging.beachball.NodalPlane(*quake['sdr0'])
    #imaging.beachball.plot_dc(np,size=20, xy=(quake['lo'],quake['la']),width=0.1)
    sdr0L.append(quake['sdr0'])
    xyL.append((quake['lo'],quake['la']))
    ball=imaging.beachball.beach(sdr0L[-1],xy=xyL[-1],width=0.1,facecolor='r',size=20)
    ax.add_collection(ball)
#plt.show()
plt.savefig('SCDYH/testAll.jpg',dpi=300)
```

```python
plt.close()
fig=plt.figure(figsize=[10,10])
ax=plt.axes()
plt.xlim(loL)
plt.ylim(laL)
ball=imaging.beachball.beach(quakes[0]['sdr0'],xy=(quakes[0]['lo'],quakes[0]['la']),width=0.5,facecolor='r',size=20)
ax.add_collection(ball)
plt.title(UTCDateTime(quakes[0]['time']))
#plt.show()
plt.savefig('SCDYH/test0.jpg',dpi=300)
```

```python
plt.close()
fig=plt.figure(figsize=[28,20])
ax=plt.axes()
plt.xlim(loL)
plt.ylim(laL)
ball=imaging.beachball.beach(quakes[-1]['sdr0'],xy=(quakes[-1]['lo'],quakes[-1]['la']),width=0.5,facecolor='r',size=20)
ax.add_collection(ball)
plt.title(UTCDateTime(quakes[-1]['time']))
plt.savefig('SCDYH/test-1.jpg',dpi=300)
```
